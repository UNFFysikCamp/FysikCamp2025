## Eulers metode
Eulers metode er en meget ligefrem algoritme der kan bruges til at finde approksimative løsninger til alle differentialligninger. Metoden udføres på følgende måde: Vi starter med en differentialligning af vilkårlig form, såsom:

$\frac{dy}{dt} = C \cdot y$

Nu ganger vi $dt$ på begge sider og får således:

$dy = C \cdot y \cdot dt$

$y_2 - y_1 = C \cdot y \cdot dt$

$y_2 = C \cdot y \cdot dt + y_1$

Det er således altid muligt at bestemme den fremtidige værdi $y_2$, når vi kender differentialligningen og den fortidige værdi $y_1$. Vi kan bruge dette til at implementere en rekursiv algoritme, der kan give os alle y-værdierne, hvis vi blot vælger en initial-værdi for $y_1$.

## Numerisk integration


Numerisk integration er meget anvendeligt, hvis løsningen er svær at finde analytisk eller slet ikke kan findes vha. almindelige funktioner. Vi kommer i løbet af denne del til at kigge på trapezmetoden og Monte Carlo integration. Der findes mange andre numeriske integrationsmetoder, men disse er intereessante at prøve at implementere selv.


Vi starter med trapezmetoden. Målet er at implementere denne numeriske integrationsmetode. Som en reminder var udtrykket for trapezmetoden givet ved:

$\int_{a}^{b} f(x) \, dx  \approx \Delta x \left( \frac{f(x_0) + f(x_N)}{2} + \sum_{i = 1}^{N-1} f(x_i)\right)$,

hvis alle intervallængderne er ens.

&emsp;

Monte Carlo integration er ikke nævnt i kompendiet, men er tænkt som en ekstra opgave, hvis man har tid. Så der er lidt meget tekst til dette.

Monte Carlo integration er en af disse metoder, hvor man udnytter evnen  til at kunne generere tilfældige tal. Ideen er at generere $N$ tilfældige punkter inden for en kasse. Arealet kan derefter estimeres ved at tælle antallet af punkter, der falder under eller over grafen.

&nbsp;

Generelt bliver udtrykket:

$ I  = \int_{\Omega}f(\mathbf{\overline{x}}) d\mathbf{\overline{x}} \approx \frac{V}{N} \sum_{i = 1}^{N}f(\mathbf{ \overline{x}_i})$,

hvor $V = \int_{\Omega} d\mathbf{\overline{x}}$ er volumen (eller længde/areal afhængigt af dimensionen) af det område, man integrerer over.

&nbsp;

Hvis man ønsker kun at integrere mellem $a$ og $b$ for en funktion af en variabel, bliver udtrykket:

$ I \approx \frac{b-a}{N} \sum_{i = 1}^{N} f(x_i) $.

## Taylor approksimationer


Nu har vi efterhånden øvet os lidt i at plotte og er klar til at bygge mere oven på.

### Teori

Polynomier er nemme at udregne, differentiere og integrere. Vi ønsker derfor nogle gange, at approksimere en ikke-polynomisk fuktion med at polynomie.

Denne approksimation kan gøres vha. Taylor-approksimationer. Først starter vi med et udtryk for approksimationen med udviklignspunkt 0:

\begin{equation}
P(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!} (x-a)^n
\end{equation}

Man kan også udvikle i andre punkter:


\begin{equation}
P(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(0)}{n!} x^n
\end{equation}
