## If statements

En af de vigtigste ting er at få computeren til at tage en beslutning. Tidligere introducerede vi "booleans" som er disse sandt/falsk statements. Vi kan give computeren et udtryk, og så tager den stilling til om det er sandt eller falsk. Så kan vi få computeren til at gøre en ting hvis det er sandt og en anden ting hvis det er falsk.

Man bruger `==`, `>`, `<`, `>=`, `<=` og `!=` til at sammenligne. `==` er om ting er ens, `>`, `<` skarpt større og mindre end, `>=`, `<=` større/mindre end eller lig med, og `!=` er ikke lig med.

```python
if x == 42:
  print("Det er det mest sande jeg nogensinde har hørt!")
```

Hvis man vil se om noget er sandt eller falsk, så kan man bruge `True` eller `False`

```python
FysikErSjovt = True

if FysikErSjovt == True:
  print("Facts!")
```

Vi kan desuden kæde dem sammen, så den tjekker flere forskellige ting. Her bruger man "elif" hvis man det første ikke gælder sig, og så kan man altid slutte af med et "else" hvis man vil samle alle dem der ikke passerer de første spørgsmål.

```python
if x > 42:
  print("Noget andet end 42? Det kan da ikke passe!")
elif x < 42:
  print("Noget mindre end 42? Nu lyver du!")
else:
  print("Alt er præcis som det skal være. Svaret er 42.")
```

## Løkker

Hvad nu hvis du virkelig bare gerne vil printe den samme tekst 100 gange? Eller måske printe alle tal fra 0 til 100 på hver sin linje. Det bliver hurtigt en RET lang kode... Men her kan computeren heldigvis hjælpe! I første omgang bruger vi `for` løkker, som gennemgår alle elementer i en liste!

En liste i python er defineret således (og kan bestå af både integers, floats og strings):

```python
liste = [1, 2.5, "fisk", 42, "funny number"]
```

Med en `for` løkke kan vi så gennemgå alle elementerne i en liste

```python
for i in liste:
  print(i)
```

Og det der med at printe alle tal fra 0 til 100? Det kan man også! (Kør på eget ansvar)

```python
for i in range(101):
  print(i)
```

Men nu tænker du da nok: "Hov - I har jo skrevet 101? Hvorfor printer den kun op til 100?" - og så svarer vi jo: Det er et mega godt spørgsmål! Det er simpelthen fordi den printer op til men ikke medtager, når man bruger `range()`.

Lille sidenote: Man kan printe individuelle elementer i en liste ved hjælp af firkantede parenteser `[]`, så hvis jeg vil have det første element f.eks.

**Bemærk:** Dataloger kan ikke tælle! De starter derfor ved 0 (nul-indeksering) hvorimod et normalt menneske ville sige at det første objekt bare er 1. Dette er vigtigt, da andet objekt i listen så er index 1 osv.

```python
print(liste[0])
```

Hvad så med det der med at printe sit eget navn 100 gange? Her kan vi bruge en anden type løkke, et såkaldt `while` loop! Det loop kører så længe dets betingelse er opfyldt. Her sammenkæder vi altså et `if` statement og en `for` løkke, i princippet! Test det selv herunder:

```python
i = 0 # Først definerer vi det variabel, som vi gerne vil loope over
navn = "Navn" # Dit (eller en andens) navn!

while i < 100: # Her definerer vi loopet, som vil køre så længe i er mindre end 100
  print("Hej mit navn er", navn)
  i += 1 # Dette betyder i bund og grund (i = i + 1) bare forkortet, så det er letlæseligt. Vi lægger 1 til i hver gang løkken kører.
```

## Funktioner

Funktioner er en nem måde at lave kode på, hvis du skal gentage den samme kode mange gange, men med forskellige input. Herunder definerer vi f.eks. en funktion der tjekker, om tallet du giver er 42.

```python
def sandhedstjekker(x):
  if x == 42:
    print("Du har fundet svaret på alting!")
  else:
    print("Din søgning må fortsætte. Dette tal er ikke rigtigt...")
```

Vi kan så kalde vores funktion med lige det input, som vi ønsker

```python
sandhedstjekker(10)
```

```python
sandhedstjekker(42)
```
